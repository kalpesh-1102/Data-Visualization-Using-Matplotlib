```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("expense_data_sample.xlsx")
df = pd.DataFrame(data)
#print(df)

group = df.groupby("Category").agg({"Amount" : "sum"})
print(group)
plt.step(group.index, group["Amount"], where= "mid", marker = "o")
plt.show()
```
