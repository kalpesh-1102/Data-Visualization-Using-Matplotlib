```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [23,43,34,56,42]

# defining the size of the figs 
plt.figure(figsize= (8,6)) # this will set the figure size of all the figures 
plt.subplot(3,2,1) # here 2 is no of rows, 2 is no of columns and 1 is the fig no 
plt.plot(x,y, ls = "--", color = "red", label = "Age", marker = "o")
plt.legend()

x = [5,6,7,8,9]
y = [45,56,57,63,82]
plt.subplot(3,2,2)
plt.scatter(x,y, color = "green", cmap= "viridis")

x = [1,3,5,7,9]
y = [45,35,56,50,70]
colors = ["red","blue","green","yellow","orange"]
plt.subplot(3,2,3)
plt.bar(x,y, color = colors, edgecolor = "black")

x = [2,4,6,8,10]
y = [45,56,30,63,45]
plt.subplot(3,2,4)
plt.stem(x,y, linefmt = "--", markerfmt= "D", bottom= 10, orientation= "horizontal")

x = [20,32,35,45,50]
plt.subplot(3,2,5)
plt.violinplot(x)

x = [1,3,5,7,9]
y = [45,35,56,50,70]
y1 = [45,56,57,63,82]
plt.subplot(3,2,6)
plt.stackplot(x,y,y1)
plt.suptitle("Employee Data") # super title 
plt.savefig("Employee Data.png") # to save the fig in the current folder
plt.show()
```
