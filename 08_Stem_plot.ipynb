```python
import matplotlib.pyplot as plt

x = [20,30,49,45,34,23,50,35]

plt.stem(x, linefmt= "--", markerfmt= "D", bottom= 10, orientation= "horizontal")
plt.show()
```

```python
import pandas as pd

data = pd.read_excel("Employee Sample Data.xlsx")
df = pd.DataFrame(data)
df2 = (df.head(50))
# print(df2)

plt.stem(df2["Age"])
plt.plot(df2["Age"])
plt.show() # this will give combination of both stem and line plot

```
