```python
import matplotlib.pyplot as plt

brands = ["Oneplus","Apple","Samsung","Nokia","Redmi"]
x = [22,35,30,3,10]
exe = [0,0,0,0,0.1]
plt.pie(x, labels = brands, explode = exe, shadow= True, autopct= "%.2f",startangle= 180)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel("expense_data_sample.xlsx")

df = pd.DataFrame(data)
print(df)
# plt.pie(df["Amount"],labels = df["Subcategory"])
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
plt.pie(grouped_by.values,labels = grouped_by.index, autopct= "%2f")
print(grouped_by)
plt.show()
```
