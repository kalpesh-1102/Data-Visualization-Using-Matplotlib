```python
import matplotlib.pyplot as plt

x = ["day1","day2","day3","day4","day5"]
y = [300,420,250,230,400]

#plt.plot(x,y)

# we can also give marker to it 
plt.plot(x,y,marker = "o", ls = "--", color = "red")
plt.show()

```

```python
x = ["day1","day2","day3","day4","day5"]
y = [300,420,250,230,400]
y1 = [500,450,350,250,320]

# comparing both the weeks visits
plt.plot(x,y,marker = "o", ls = "--", color = "red", label = "Week 1")
plt.plot(x,y1,marker = "o", ls = "-", color = "green", label = "Week 2", alpha = 0.5) # alpha is transperancy, lies betn 0-1
plt.legend()
plt.xlabel("Days of visit",fontsize = "10")
plt.ylabel("No of people visited",fontsize = "10")
plt.show()
```
