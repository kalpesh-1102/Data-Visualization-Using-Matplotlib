```python
import pandas as pd

data = pd.read_excel("Employee Sample Data.xlsx")
df = pd.DataFrame(data)
# print(df)
plt.violinplot(df["Age"], showmedians= True)
plt.xlabel("Employee")
plt.ylabel("Age")
plt.show()
```
